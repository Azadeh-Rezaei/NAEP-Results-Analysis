```python
# create a subset of the dataframe df
df_clean = df[var_list]
```

```python
df_clean.dtypes
```

```python
# Set to show all the columns
pd.set_option('display.max_rows', None)
```

```python
# Finding the number of missing values
df_clean.isnull().sum()
```

```python
df_clean = df_clean.drop(df_clean[df_clean['STATE'].isin(['NATIONAL', 'DODEA'])].index)
```

```python
# Only keeping the data for the years 2003 and after
df_clean = df_clean[df_clean['YEAR'] > 2002 ]
```

```python
# Resetting the index:
df_clean = df_clean.reset_index()
```

```python
# Deleting the old index columns:
del df_clean['index']
```
